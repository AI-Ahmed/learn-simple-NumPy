### Binning
---
suppose a group of 12 sales price records has been sorted as follows:

        5, 10, 11, 13, 15, 35, 50, 55, 72, 92, 204, 215.

- Partition them into three bins by each of the following method:

        - Equal-width Partitioning


```python
import pandas as pd
import numpy as np

df = pd.DataFrame([5, 10, 11, 13, 15, 35, 50, 55, 72, 92, 204, 215], columns=['price'])
bins = np.linspace(df['price'].min(), df['price'].max(), 4)
group_names = ['Low', 'Medium', 'Large']
df['binning'] = pd.cut(df['price'], bins, labels=group_names, include_lowest=True)

df
```

```python
### ----- Implement that into plot ----- ###
import matplotlib.pyplot as plt

plt.plot(df['binning'].tolist(), df['price'], label='list of price')
plt.legend()
plt.xlabel('binning')
plt.ylabel('price')
```

```python
### ----- Implement that into histogram ----- ###
import matplotlib.pyplot as plt

plt.bar(df['binning'].tolist(), df['price'])
plt.xlabel('Binned Items')
plt.ylabel('Price')
plt.title('List of price')
```

What's the right plot to use ?

        If you notice, histogram plot doesn't exactly show the right answer, it doesn't show the amount, it shows the frequencey of the numbers not their amount.

        Unlike the freq plot which it shows the curve of the numbers

You can also show the number of each bins

```python
import matplotlib.pyplot as plt

plt.hist(df['binning'].tolist(), bins=3)
plt.xlabel('Binned Items')
plt.ylabel('Count')
plt.title('List of price')
```

This's the right answer!
